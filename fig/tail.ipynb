```python
import numpy as np # Linear Algebra
from matplotlib import pyplot as plt # Plotting
plt.rcParams['font.family'] = 'serif'
plt.rcParams["font.size"] = 16
```

```python
beta = 1.0
N = 1000

taus = np.linspace(0, beta, N)
taus_ex = np.hstack((taus - beta, taus, taus + beta))
```

```python
g1 = np.full(N, -0.5)
g2 = 0.25 * (-beta + 2 * taus)
g3 = 0.25 * (beta * taus - taus**2)
```

```python
def antip(g):
    return np.hstack((-g, g, -g))
```

```python
g1_ex = antip(g1)
g2_ex = antip(g2)
g3_ex = antip(g3)
```

```python
colors = ["r", "g", "b"]

for shift, s in zip([-beta, 0, beta], [-1, 1, -1]):
    plt.plot(taus+shift, s * g1, color=colors[0])
    plt.plot(taus+shift, s * g2, color=colors[1])
    plt.plot(taus+shift, s * g3, color=colors[2])
plt.grid()
plt.xlabel(r"$\tau$")
plt.savefig("tail.png", bbox_inches="tight")
```
